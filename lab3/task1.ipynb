```python
import pandas as pd
```

Таблица рейтингов

```python
ratings_dataframe = pd.read_csv("../lab3/data/ratings.dat", sep="::", names=["UserID", "MovieID", "Rating", "Timestamp"], engine="python")
ratings_dataframe
```

Таблица пользователей

```python
users_dataframe = pd.read_csv("../lab3/data/users.dat", sep="::", names=["UserID", "Gender", "Age", "Occupation", "Zip-code"], engine="python")
users_dataframe
```

Таблица фильмов

```python
movies_dataframe = pd.read_csv("../lab3/data/movies.dat", sep="::", names=["MovieID", "Title", "Genre"], engine="python", encoding="latin-1")
movies_dataframe
```

1) Требуется вычислить средние рейтинги для фильма с ID = № варианта в зависимости от пола;

```python
variant = 48
user_rating_merge_dataframe = pd.merge(users_dataframe, ratings_dataframe, on="UserID")
user_rating_var_dataframe = user_rating_merge_dataframe[user_rating_merge_dataframe["MovieID"] == variant]
female_ratings = user_rating_var_dataframe[user_rating_var_dataframe["Gender"] == "F"]['Rating'].mean()
male_ratings = user_rating_var_dataframe[user_rating_var_dataframe["Gender"] == "M"]['Rating'].mean()
print(f"Средний рейтинг для фильма среди женщин - {round(female_ratings, 2)}\nСредний рейтинг для фильма среди мужчин - {round(male_ratings, 2)}")
movies_dataframe[movies_dataframe["MovieID"] == variant]
```

2) Отобрать фильмы, получившие не менее 250 оценок, и для них определить средний рейтинг в зависимости от пола;

Результат вывести в виде DataFrame с колонками:

['MovieID', 'Title', 'Rating_F', 'Rating_M']

```python
general_dataframe = pd.merge(user_rating_merge_dataframe, movies_dataframe, on="MovieID")

most_rated = general_dataframe.groupby("MovieID").size()
most_rated = most_rated[most_rated >= 250]

most_rated_dataframe = general_dataframe[general_dataframe["MovieID"].isin(most_rated.index)]
ratings = round(most_rated_dataframe.groupby(["MovieID", "Gender"])['Rating'].mean(), 2).unstack()
ratings_dataframe = pd.merge(ratings, movies_dataframe[["MovieID", "Title"]], on="MovieID")
ratings_dataframe = ratings_dataframe.rename(columns={"F": "Rating_F", "M": "Rating_M"})
ratings_dataframe = ratings_dataframe.reindex(columns=["MovieID", "Title", "Rating_F", "Rating_M"])

display(most_rated_dataframe)
display(ratings_dataframe)
display(general_dataframe)
```

3) Найти фильмы, оказавшиеся в топ 5 у зрителей-женщин и худшие 5 у зрителей- мужчин. Найти фильмы (топ 5), по которым мужчины и женщины сильнее всего разошлись в оценках;

```python
top_five_F = ratings_dataframe.sort_values(by="Rating_F", ascending=False)[:5]
display(top_five_F)

worst_five_M = ratings_dataframe.sort_values(by="Rating_M", ascending=True)[:5]
display(worst_five_M)

ratings_dataframe["Difference"] = ratings_dataframe.apply(lambda x: abs(x["Rating_F"] - x["Rating_M"]), axis=1)
top_five_dif = ratings_dataframe.sort_values(by="Difference", ascending=False)[:5]
display(top_five_dif)
```

4) Определить фильмы (топ 5), вызвавшие в среднем наибольшее разногласие у зрителей независимо от пола;

```python
rating_std = general_dataframe.groupby("MovieID")["Rating"].std().round(2)
rating_std.name = "STD"
rating_std_dataframe = pd.merge(rating_std, movies_dataframe[["MovieID", "Title"]], on="MovieID")
rating_std_dataframe = rating_std_dataframe.reindex(columns=["MovieID", "Title", "STD"])

rating_std_dataframe = rating_std_dataframe.sort_values("STD", ascending=False).head(5)
display(rating_std_dataframe)
```

5) Определить средний рейтинг каждого жанра для различных возрастов. Установите, фильмы какого жанра в каждой категории возрастов имеют наибольший и наименьший рейтинги. Если фильм относится к нескольким данным, то их надо разделить и сформировать несколько строчек с этим фильмом, но с различными жанрами, например:

| UserID<br> | MovieID | Rating | Timestamp | Gender | Age | Occupation | Zip-code |              Title               |               Genres               |

необходимо преобразовать в

| UserID | MovieID | Rating | Timestamp | Gender | Age | Occupation | Zip-code |              Title               |   Genres   |
| :----: | :-----: | :----: | :-------: | :----: | :-: | :--------: | :------: | :------------------------------: | :--------: |

```python
general_dataframe["Genre"] = general_dataframe["Genre"].str.split("|")
exploded_general_dataframe = general_dataframe.explode("Genre")
genre_ratings = round(exploded_general_dataframe.groupby(["Genre", "Age"])["Rating"].mean(), 2).reset_index()
print("Средняя оценка жанров по возрасту")
display(genre_ratings)

max_rating_by_age = genre_ratings.loc[genre_ratings.groupby("Age")["Rating"].idxmax()]
min_rating_by_age = genre_ratings.loc[genre_ratings.groupby("Age")["Rating"].idxmin()]

print("Максимальные оценки по возрастам:")
display(max_rating_by_age)

print("Минимальные оценки по возрастам:")
display(min_rating_by_age)
```

### задание 1 для защиты
вывести фильмы, которые мужчины смотрели больше в 2 раза, чем женщины

```python
counts = general_dataframe.groupby(['MovieID', 'Gender']).size()
print(counts)
counts_unstacked = counts.unstack()
print(counts_unstacked)

male_double_females_df = counts_unstacked[counts_unstacked['M'] >= 2 * counts_unstacked['F']]
display(male_double_females_df)

result_final_df = pd.merge(male_double_females_df, movies_dataframe[['MovieID', 'Title']], on='MovieID')

result_final_df = result_final_df.sort_values(by='MovieID')
display(result_final_df)
```
