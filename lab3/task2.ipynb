```python
from scipy.stats import norm
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt
```

1) Случайная величина $X$ имеет нормальное распределение, $M[X] = 10$ и $D[X] = 4$

```python
norm_m = 10
norm_std = 2
normal_dist = norm(norm_m, norm_std)

norm_normal_dist = norm(0, 1)
```

- Построить функцию и плотность распределения вероятностей для нормированной нормально распределенной случайной величины.

```python
norm_x = np.linspace(-5, 5, 100)

norm_pdf = norm_normal_dist.pdf(norm_x)

norm_cdf = norm_normal_dist.cdf(norm_x)

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(norm_x, norm_cdf, 'r')
plt.title('Функция распределения\nНормированное нормальное распределение')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(norm_x, norm_pdf, 'b')
plt.title('Плотность распределения вероятностей\nНормированное нормальное распределение')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)

plt.show()
```

- Построить функцию и плотность распределения вероятностей для $X$.

```python
norm_x = np.linspace(0, 20, 100)

norm_pdf = normal_dist.pdf(norm_x)

norm_cdf = normal_dist.cdf(norm_x)

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(norm_x, norm_cdf, 'r')
plt.title('Функция распределения вероятностей\nНормальное распределение')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(norm_x, norm_pdf, 'b')
plt.title('Плотность распределения вероятностей\nНормальное распределение')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)


plt.show()
```

- Вычислить вероятности $P(X < 11)$, $P(X > 5)$, $P(4 < X < 12)$.


```python
norm_P1 = normal_dist.cdf(11) # P(X < 11)

norm_P2 = 1 - normal_dist.cdf(5) # P(X > 5)

norm_P3 = normal_dist.cdf(12) - normal_dist.cdf(4) # P(4 < X < 12)

print(f"P(X < 11) = {round(norm_P1, 4)}")
print(f"P(X > 5) = {round(norm_P2, 4)}")
print(f"P(4 < X < 12) = {round(norm_P3, 4)}")
```

- Отобразить, вероятности на графиках функции и плотности вероятностей.

```python
plt.figure(figsize=(15, 15))

plt.subplot(3, 2, 1)
plt.plot(norm_x, norm_cdf, 'r', linewidth=2)
plt.axhline(y=norm_P1, color='red', linestyle='--', alpha=0.7)
plt.title('P(X < 11). Функция распределения вероятностей')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 2)
plt.plot(norm_x, norm_pdf, 'b')
plt.fill_between(norm_x[norm_x < 11], norm_pdf[norm_x < 11], alpha=0.3, color='blue')
plt.title('P(X < 11). Плотность распределения вероятностей')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 3)
plt.plot(norm_x, norm_cdf, 'r', linewidth=2)
plt.axhline(y=1, color='red', linestyle='--', alpha=0.7)
plt.axhline(y=normal_dist.cdf(5), color='red', linestyle='--', alpha=0.7)
plt.title('P(X > 5). Функция распределения вероятностей')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 4)
plt.plot(norm_x, norm_pdf, 'b')
plt.fill_between(norm_x[norm_x > 5], 0, norm_pdf[norm_x > 5], alpha=0.3, color='blue')
plt.title('P(X > 5). Плотность распределения вероятностей')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 5)
plt.plot(norm_x, norm_cdf, 'r', linewidth=2)
plt.axhline(y=normal_dist.cdf(4), color='red', linestyle='--', alpha=0.7)
plt.axhline(y=normal_dist.cdf(12), color='red', linestyle='--', alpha=0.7)
plt.title('P(4 < X < 12). Функция распределения вероятностей')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 6)
plt.plot(norm_x, norm_pdf, 'b')
plt.fill_between(norm_x[(norm_x > 4) & (norm_x < 12)], 0, norm_pdf[(norm_x > 4) & (norm_x < 12)], alpha=0.3, color='blue')
plt.title('P(4 < X < 12). Плотность распределения вероятностей')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)

plt.show()
```

- Проверить выполнение правила «3-сигм» для $X$, отобразить это правило на графике плотности вероятностей

```python
P_std3 = normal_dist.cdf(norm_m + 3*norm_std) - normal_dist.cdf(norm_m - 3*norm_std)

print(f"P(x ∈ [m - 3σ, m + 3σ] = {round(P_std3, 4)}")

down = norm_m - 3*norm_std
up = norm_m + 3*norm_std

plt.figure(figsize=(10, 6))
plt.plot(norm_x, norm_pdf, 'b-', linewidth=2, label='Плотность вероятности')
mask_3_sigma = (norm_x >= down) & (norm_x <= up)
plt.fill_between(norm_x[mask_3_sigma], norm_pdf[mask_3_sigma], alpha=0.3, color='blue')
plt.axvline(down, color='red', linestyle='--', alpha=0.7)
plt.axvline(up, color='red', linestyle='--', alpha=0.7)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Правило 3-сигм')
plt.grid(True, alpha=0.3)
plt.show()
```

2) Случайная величина $Y$ имеет экспоненциальное распределение, $M[Y] = 2$

```python
exp_m = 2
exp_dist = expon(scale=exp_m)
```

- Построить для нее функцию и плотность распределения вероятностей.

```python
y = np.linspace(0, 10, 100)

exp_pdf = exp_dist.pdf(y)

exp_cdf = exp_dist.cdf(y)

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(y, exp_cdf, 'r')
plt.title('Функция распределения вероятностей\nЭкспоненциальное распределение')
plt.xlabel('y')
plt.ylabel('F(y)')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(y, exp_pdf, 'b')
plt.title('Плотность распределения вероятностей\nЭкспоненциальное распределение')
plt.xlabel('y')
plt.ylabel('f(y)')
plt.grid(True, alpha=0.3)

plt.show()
```

- Вычислить вероятности $P(Y < 3)$, $P(Y > 2)$, $P(1 < Y < 2.5)$.


```python
exp_P1 = exp_dist.cdf(3) # P(Y < 3)

exp_P2 = 1 - exp_dist.cdf(2) # P(Y > 2)

exp_P3 = exp_dist.cdf(2.5) - exp_dist.cdf(1) # P(1 < Y < 2.5)

print(f"P(Y < 3) = {round(exp_P1, 4)}")
print(f"P(Y > 2) = {round(exp_P2, 4)}")
print(f"P(1 < Y < 2.5) = {round(exp_P3, 4)}")
```

- Отобразить, вероятности на графиках функции и плотности вероятностей.

```python
plt.figure(figsize=(15, 15))

plt.subplot(3, 2, 1)
plt.plot(y, exp_cdf, 'r', linewidth=2)
plt.axhline(y=exp_P1, color='red', linestyle='--', alpha=0.7)
plt.title('P(Y < 3). Функция распределения')
plt.xlabel('y')
plt.ylabel('F(y)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 2)
plt.plot(y, exp_pdf, 'b', linewidth=2)
plt.fill_between(y[y < 3], exp_pdf[y < 3], alpha=0.3, color='blue')
plt.title('P(Y < 3). Плотность распределения')
plt.xlabel('y')
plt.ylabel('f(y)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 3)
plt.plot(y, exp_cdf, 'r', linewidth=2)
plt.axhline(y=1, color='red', linestyle='--', alpha=0.7)
plt.axhline(y=exp_dist.cdf(2), color='red', linestyle='--', alpha=0.7)
plt.title('P(Y > 2). Функция распределения')
plt.xlabel('y')
plt.ylabel('F(y)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 4)
plt.plot(y, exp_pdf, 'b', linewidth=2)
plt.fill_between(y[y > 2], 0, exp_pdf[y > 2], alpha=0.3, color='blue')
plt.title('P(Y > 2). Плотность распределения')
plt.xlabel('y')
plt.ylabel('f(y)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 5)
plt.plot(y, exp_cdf, 'r', linewidth=2)
plt.axhline(y=exp_dist.cdf(1), color='red', linestyle='--', alpha=0.7)
plt.axhline(y=exp_dist.cdf(2.5), color='red', linestyle='--', alpha=0.7)
plt.title('P(1 < Y < 2.5). Функция распределения')
plt.xlabel('y')
plt.ylabel('F(y)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 6)
plt.plot(y, exp_pdf, 'b', linewidth=2)
plt.fill_between(y[(y > 1) & (y < 2.5)], 0, exp_pdf[(y > 1) & (y < 2.5)], alpha=0.3, color='blue')
plt.title('P(1 < Y < 2.5). Плотность распределения')
plt.xlabel('y')
plt.ylabel('f(y)')
plt.grid(True, alpha=0.3)

plt.show()
```

- Проверить выполнение правила «3-сигм» для $Y$ , отобразить это правило на графике плотности вероятностей.

```python
p_y_3_sigma = exp_dist.cdf(exp_m + 3 * exp_m)

plt.plot(y, exp_pdf, 'b-', linewidth=2, label='Плотность вероятности')
mask_y_3_sigma = y <= exp_m + 3 * exp_m
plt.fill_between(y[mask_y_3_sigma], exp_pdf[mask_y_3_sigma], alpha=0.3, color='blue')
plt.axvline(exp_m + 3 * exp_m, color='red', linestyle='--', alpha=0.7)
plt.title('Правило "3-сигм"', fontsize=14, fontweight='bold')
plt.xlabel('y', fontsize=12)
plt.ylabel('Вероятность', fontsize=12)
plt.grid(True, alpha=0.3)
print(p_y_3_sigma)
```
