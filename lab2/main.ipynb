```python
import pandas as pd
from datetime import datetime
```

```python
dataframe = pd.read_csv("../lab2/data/BikeData.csv")
dataframe
```

1) Создайте новую переменную Total_count, в которую записать общее количество арендованных велосипедов за каждый час.

```python
dataframe['Total_count'] = dataframe['Partner 1'] + dataframe['Partner 2']
dataframe
```

2) Удалите ряды Partner 1 и Partner 2

```python
dataframe = dataframe.drop(columns=['Partner 1', 'Partner 2'])
dataframe
```

3) Преобразуйте переменную Date к стандартной форме (to_datetime).

```python
# dataframe['Date'] = dataframe['Date'].dt.strftime('%d.%m.%Y')
dataframe['Date'] = pd.to_datetime(dataframe['Date'], format='%d/%m/%Y')
dataframe
```

4) Определите сколько велосипедов было выдано в день, месяц и число которого совпадает с днем выполнения этого задания. Определите, какой это был день недели.

```python
current_date = datetime.now().date()
current_month = current_date.month
current_day = current_date.day

current_date_dataframe = dataframe[(dataframe['Date'].dt.month == current_month) & (dataframe['Date'].dt.day == current_day)]
current_date_count = current_date_dataframe['Total_count'].sum()
table_day = current_date_dataframe['Date'].dt.weekday.iloc[0]
display(current_date_dataframe)
print(f"Количество: {current_date_count}\nДень недели: {table_day}")
```

5) Перекодируйте категориальную переменную переменную Functioning Day(рабочий или нерабочий день) в переменную типа boolean, т.е. "Yes"=True и "No" = False. Выведите количество нерабочих дней ( не часов ). День считается нерабочим, если пункт выдачи велосипедов работал менее 12 часов.

```python
def get_func_day_state(state):
    if state == 'Yes':
        return True
    else:
        return False

dataframe['Functioning Day'] = dataframe['Functioning Day'].apply(get_func_day_state)
display(dataframe)

functioning_hours = dataframe.groupby(dataframe['Date'])['Functioning Day'].sum()
display(functioning_hours)
not_functioning_days_count = (functioning_hours < 12).sum()

print(f"Количество нерабочих дней: {not_functioning_days_count}")
```

6) Перекодируйте переменную Holiday (являлся ли день праздничным) в 0, если No Holiday, и 1, если Holiday. (используйте анонимную функцию lambda).

```python
dataframe['Holiday'] = dataframe['Holiday'].apply(lambda x: 1 if x == 'Holiday' else 0)
display(dataframe)
```

7) Введите новую категориальную переменную Temperature category, которая будет равна:
- "Freezing", если температура < 0;
- "Chilly", если 0 <= температура < 15;
- "Nice", если 15 <= температура < 26;
- "Hot", если 26 <= температура.

```python
def set_temperature_category(temperature):
    if temperature < 0:
        return "Freezing"
    elif (temperature >= 0) & (temperature < 15 ):
        return "Chilly"
    elif (temperature >= 15) & (temperature < 26):
        return "Nice"
    elif (temperature >= 26):
        return "Hot"

dataframe['Temperature Category'] = dataframe['Temperature'].apply(set_temperature_category)
dataframe
```

8) Создайте новую переменную Good weather, которая будет равна 1 (т.е. хорошая погода), если: Temperature category = "Nice", Humidity в диапазоне от 40 до 60 , скорость ветра Wind speed меньше 5.4, нет дождя (Rainfall) нет снега (Snowfall). Определите, сколько процентов наблюдений соответствует хорошей погоде.


```python
dataframe['Good weather'] = dataframe.apply(lambda x: 1 if (x['Temperature Category'] == 'Nice') & (x['Humidity'] >= 40) & (x['Humidity'] <= 60) & (x['Wind speed'] < 5.4) & (x['Rainfall'] == 0.0) & (x['Snowfall'] == 0.0) else 0, axis=1)

good_weather_percent = (dataframe[dataframe['Good weather'] == 1].shape[0] / dataframe.shape[0]) * 100
display(dataframe)
print(f"Хорошая погода - {round(good_weather_percent, 2)} %")
```

9) Сгруппируйте данные по сезонам (Seasons), внутри сезона по категории погоды (Temperature category) и вывести для полученной группировки количество арендованных велосипедов. Вывести результаты в виде сгруппированной таблицы (через groupby) и в виде сводной таблицы (pivot_table).

```python
grouped = dataframe.groupby(['Seasons', 'Temperature Category'])['Total_count'].sum()

pivot = dataframe.pivot_table(
    index='Seasons',
    columns='Temperature Category',
    values='Total_count',
    aggfunc='sum'
)

display(grouped)
display(pivot)
```
